```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv(r"/content/drive/MyDrive/Deep-Learning Youtube/Airline_Delay_Cause.csv")
```

```python
data
```

```python
data.info()
```

```python
data = data.drop(['carrier','carrier_name','airport','airport_name'],axis=1)
data
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data['weather_delay'].min(),data['weather_delay'].max()
```

```python
data['WDCase'] = data['weather_delay'].apply(lambda x : 1 if x> 100 else 0 )
```

```python
data['WDCase'].value_counts()
```

```python
data
```

```python
x = data.drop(['WDCase'],axis =1)
y = data['WDCase']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size=0.25 , shuffle=True)

print("x_train shape : ",x_train.shape)
print("x_test shape : ",x_test.shape)
print("y_train shape : ",y_train.shape)
print("y_test shape : ",y_test.shape)
```

```python
import tensorflow as tf
import keras
```

```python
kerasModel = keras.models.Sequential([
    #  keras.layers.Input(shape=(17))
    keras.layers.Dense(8,activation='sigmoid'),
    # keras.layers.Dropout(0.2),  for overfitting
    keras.layers.Dense(128,activation='tanh'),
    keras.layers.Dense(64,activation='tanh'),
    # keras.layers.Dropout(0.2),  for overfitting
    keras.layers.Dense(1,activation='sigmoid')

    ])
```

```python
MyOptimizer= tf.keras.optimizers.AdamW(
    learning_rate=0.001,
    weight_decay=0.0001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name="AdamW"

)
```

```python
kerasModel.compile(optimizer=MyOptimizer,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = kerasModel.fit(x_train,y_train,
                         validation_data=(x_test,y_test),
                         epochs=100,
                         batch_size = 10000,
                         verbose=1,
                         callbacks= [tf.keras.callbacks.EarlyStopping(
                             patience = 10,
                             monitor="val_accuracy",
                             restore_best_weights=True
                         )])
```

```python
print(kerasModel.summary())
```

```python
kerasModel.save('kerasModel.keras')
```

```python
NewKerasModel = keras.models.load_model('kerasModel.keras')
```

```python
y_pred = NewKerasModel.predict(x_test)
```

```python
print("Prediction shape is " ,y_pred.shape)
```

```python
print("prediction items are",y_pred[:10])
```

```python
y_test[:10]
```

```python
ModelLoss , ModelAccuracy = NewKerasModel.evaluate(x_test,y_test)
print("Model loss is ",ModelLoss)
print("Model Accuracy is ",ModelAccuracy)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
